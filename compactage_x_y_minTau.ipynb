### 1.3. Formalisme mathématique

#### 1.3.1. Ensembles principaux

Afin de formellement décrire l'ensemble des passagers de l'avion ainsi que les sièges de la cabine, introduisons les ensembles suivants :
<ul>
    <li> $\mathcal{P}$ l'ensemble des passagers ;
    <li> $\mathcal{S}_\emptyset$ l'ensemble des sièges vides ; 
    <li> $\mathcal{S} = [1,28] \times [1,7] \setminus \mathcal{S}_\emptyset$ l'ensemble des sièges réels.   
</ul>
A noter que si on voulait exprimer exhaustivement l'ensemble des sièges vides, on aurait donc : 
<br>
$$\mathcal{S}_\emptyset = ([ 1,28 ] \times \lbrace 4 \rbrace) \cup (\lbrace 1 \rbrace \times [ 5,8 ])$$

#### 1.3.2. Variables indicatrices

A partir de ces ensembles, on peut donc introduire les variables indicatrices $\delta_{xyp} \in \lbrace 0,1 \rbrace$ $(p \in \mathcal{P}, (x,y) \in \mathcal{S})$ valant $1$ si le passager $p$ est assis sur le siège de coordonnées $(x,y)$, et $0$ sinon.

Les variables $\delta_{xyp}$ nous serons très utiles pour exprimer des conditions liées à l'autorisation ou non que des passagers s'assoient sur un siège. Les conditions de base sont :
<ul>
    <li> Au plus un passager par siège : $\forall (x,y) \in \mathcal{S}$, $\sum_{p \in \mathcal{P}} \delta_{xyp} \leq 1$ ;
    <li> Au plus un siège par passager : $\forall p \in \mathcal{P}$, $\sum_{(x,y) \in \mathcal{S}} \delta_{xyp} \leq 1$ ;
    <li> Tous les passagers ont un siège : $\sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}} \delta_{xyp} = \# \mathcal{P}$ ;   
</ul>

Ce modèle sera enrichi par la suite avec les différentes contraintes qui s'ajouteront à notre modèle. Par exemple, un détail important sur lequel nous nous attarderont subséquemment est le fait qu'aucun passager enfant ne doit être assis sur une rangée située au niveau des issues de secours de l'avion. Sans s'épancher sur la définition formel des deux ensemble suivants, nous pouvons écrire :
<ul>
    <li> $\mathcal{E} \subset \mathcal{P}$ l'ensemble des passagers enfants ;
    <li> $\mathcal{S}_\mathcal{I} \subset \mathcal{S}$ l'ensemble des sièges situés au niveau des issues de secours.
</ul>
Dès lors, on aura : $\sum_{p \in \mathcal{E}} \sum_{(x,y) \in \mathcal{S}_\mathcal{I}} \delta_{xyp} = 0$, ce qui signifie bien qu'aucun enfant ne devra être assis près d'une issue de secours. Toutefois, cette condition ne sera envisagée que dans la seconde phase de notre étude donc nous y reviendrons ultérieurement.
<br> <br>
Cette partie avait juste pour but d'introduire formellement quelques ensembles d'intérêt ainsi que les variable $\delta_{xyp}$ qui nous serons très utiles par la suite, notamment en tant que variable de décision pour le problème d'optimisation.

#### Agglomération des variables indicatrices individuelles en variables indicatrices de groupe

Pour les groupes de plus s'un passager, il peut être intéressant d'introduire des variables indicatrices équivalentes aux $\delta_{xyp}$ mais à l'échelle du groupe. Ainsi, si on appelle $\mathcal{G}$ l'ensemble des groupes de passagers et $\mathcal{G}_{2}^{+}$ l'ensemble des groupes de plus d'au moins deux passagers, on a :
<br><br>
$$\forall (x,y) \in \mathcal{S}, \forall g \in \mathcal{G}_{2}^{+} : \delta_{xyg} = \sum_{p \in g} \delta_{xyp}$$

Ainsi, la variable $\delta_{xyg}$ vaut $1$ si et seulement si un membre du groupe $g$ est assis en $(x,y)$. De telles variables seront très utiles par la suite car elles entreront en compte dans la formulation de propriétés vérifiées de manière équivalentes par l'ensemble des membres d'un même groupe.

## 2. Contraintes liées à la sécurité

### 2.1. Centrage de l'avion

Afin que l'avion puisse décoller sans risque et qu'il reste stable une fois en vol, le barycentre ce de dernier doit se situer dans une zone très particulière appelée "zone de centrage". Il s'agit d'une contrainte de sécurité forte devant être respectée obligatoirement. Cette zone se situe entre la $13^{ème}$ et la $17^{ème}$ rangée dans le sens de la longeur, et entre les deux colonnes bordant l'allée centrale dans le sens de la hauteur.

Dans le cadre de notre problème, on considère que le barycentre de l'avion à vide se situe effectivement dans cette zone. Il nous faut donc vérifier que la répartition des masses due au positionnement des passagers conserve le barycentre dans cette même zone de centrage.

#### 2.1.1. Formalisme mathématique

Considérons un ensemble de $N$ masses $(m_i)_{i \in [1,N]}$ réparties dans le plan. Pour chaque masse $m_i$, on désigne par $\vec{p_i}=(x_i,y_i)^T$ son vecteur position en partant de l'origine dudit plan. Dès lors, le vecteur position du barycentre $\vec{B}$ se calcule par :
$$\vec{B}=\frac{\sum_{i=1}^N m_i\vec{p_i}}{\sum_{i=1}^N m_i}$$
Dès lors, en projetant cette relation selon la direction $x$ et la direction $y$, on obtient deux équations découplées donnant la position barycentre :
<ul>
    <li> Positionnement selon $x$ : $B_x=\frac{\sum_{i=1}^N m_i x_i}{\sum_{i=1}^N m_i}$
    <li> Positionnement selon $y$ : $B_y=\frac{\sum_{i=1}^N m_i y_i}{\sum_{i=1}^N m_i}$
</ul>        

#### 2.1.2. Expression à l'aide des $\delta_{xyp}$

Compte tenu des définitions précédentes, on peut alors réécrire les conditions à l'aide des variables $\delta_{xyp}$ en considérant la direction $x$ définie ci-dessus comme la direction dans la longueur de l'avion, et $y$ celle dans la hauteur de ce dernier :
<ul>
    <li> Centrage selon la longueur : $13 \sum_{p \in \mathcal{P}} m_p \leq \sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}} m_p x \delta_{xyp} \leq 17 \sum_{p \in \mathcal{P}} m_p$
    <li> Centrage selon la largeur : $3 \sum_{p \in \mathcal{P}} m_p \leq \sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}} m_p y \delta_{xyp} \leq 5 \sum_{p \in \mathcal{P}} m_p$  
</ul>

### 2.2. Temps de correspondance

Le temps que les passagers mettent pour sortir de l'avion est d'autant plus court qu'ils sont situés à l'avant de ce dernier. Pour les passagers en correspondance, sortir rapidement de la cabine n'est pas qu'un simple argument de satisfaction, il s'agit d'une réelle nécessité. Dans cette premièr instance de notre solution au problème, nous avons fait le choix d'ajouter une contrainte liant la rangée du siège des passagers à leur temps de correspondance. Ainsi, les deux propriétés devront être satisfaites :
<ul>
    <li> Si le passager a un temps de correspondance de moins de 90 minutes, il devra nécessairement être assis dans moitié avant de l'avion;
    <li> Si le passager a un temps de correspondance de moins de 120 minutes, il devra nécessairement être assis dans les deux tiers avant de l'avion.
</ul>
Nous considérons ici que si un passager possède un temps de correspondance supérieur à deux heures, sa place dans l'avion n'affecte pas significativement le risque qu'il a de rater son avion suivant

Mathématiquement, introduisons les ensembles $\mathcal{P}_{i}^{\tau}$ représentant l'ensemble des passagers ayant un temps de correspondance inférieur à $i$ minutes.

Avec le formalisme des $\delta_{xyp}$, ces contraintes deviennent :
<ul>
    <li> $\forall p \in \mathcal{P}_{90}^{\tau}$, $\sum_{(x,y) \in \mathcal{S}} x \delta_{xyp} \leq 15$
    <li> $\forall p \in \mathcal{P}_{120}^{\tau}$, $\sum_{(x,y) \in \mathcal{S}} x \delta_{xyp} \leq 20$
</ul>

## 3. Fonction à maximiser

Dans le cadre de ce problème, on cherche à maximiser le contentement des passagers d'un point de vue confort et pratique Pour cela, on cherche donc à faire en sorte que les passagers d'un même groupe soient assis à côté

### 3.1. Définition du voisinage

On considère que le voisinage d'un siège est constitué des deux sièges à côté de lui, ainsi que des sièges devant et derrière lui dans le cas général. Lorsqu'il s'agit d'un siège à côté du couloir central, on considère que le siège de la même rangée de l'autre côté du couloir est aussi dans le voisinage.

Toutefois, il est évident que deux passagers du même groupe seront plus satisfaits s'ils sont côte-à-côte plutôt que séparés par un couloir ou l'un devant l'autre. Pour cela, on introduit différents coeffcients de contentement relatifs à la position relative de deux passagers, dont le plus grand est bien évidemment celui correspondant à la situation dans laquelle les passagers sont côte-à-côte :
<ul>
    <li> $c_{côté}$ lorsque les passagers sont côte-à-côte;
    <li> $c_{derrière}$ lorsqu'un des deux passagers est assis derrière l'autre;
    <li> $c_{couloir}$ lorsque les passagers sont séparés par le couloir central.
</ul>

### 3.2. Introduction de nouvelles variables de décision

On introduit donc, pour <i>mesurer</i> le score de contentement d'un passager vis-à-vis de son voisinnage, des variables $s_{xyp}$. Pour illustrer le fonctionnement de ces variables, donnons l'exemple d'un passager $p$ appartenant au groupe $g \in \mathcal{G}_{2}^{+}$. Alors, on a : 
<br><br>
$$s_{xyp} = \sum_{q \in G \setminus \lbrace p \rbrace} c_{côté} ( \delta_{x,y+1,q} + \delta_{x,y-1,q}) 
+ c_{devant} (\delta_{x+1,y,q} + \delta_{x-1,y,q})$$
<br>
Avec des mots : "on calcule le score par passager en sommant de manière pondérée la présence ou non de membres du même groupe dans le voisinnage du passager".
<br><br>
A noter qu'en l'abscence de passager sur le siège de coordonnées $(x,y)$, la valeur de $s_{xyp}$ est nécessairement nulle pour tout $p$. Formellement, $\delta_{xyp} = 0 \Rightarrow s_{xyp} = 0$.

### 3.3. Analyse des différents cas

#### 3.3.1. Siège situé côté couloir

Pour les sièges donnant sur le couloir, on rajoute un terme à la somme définissant $s_{xyp}$ et correspondant au poids de la présence d'un membre du groupe de l'autre côté du couloir. Ceci se produit donc pour les sièges tels que $y \in \lbrace 3,5 \rbrace$.

#### 3.3.2. Distinction sur la taille du groupe

Quand le groupe est constitué de trois personnes ou moins, idéalement, tout le groupe doit tenir sur trois sièges juxtaposés sur la même rangée. Dès lors, il serait contre-productif de favoriser le fait que les membres du groupe soient placés sur différentes rangées ou de part et d'autre du couloir. Ainsi, les coefficients $c_{devant}, c_{diagonale},$ et $c_{couloir}$ auront des valeurs différentes selon que le cardinal du groupe considéré est inférieur ou égal à trois, ou strictement supérieur à cette valeur (ces trois constantes étant vraisemblablement nulles ou très petites dans le cas des groupes à moins de trois membres).

### 3.4. Raffinement du modèle

#### 3.4.1. Equivalence entre les membres d'un même groupe

On peut tout d'abord remarquer que le score de voisinage est le même quelque soit le membre du groupe considéré. Dès lors, nous pouvons considérablement réduire les variables de score en ne les définissant que pour les groupes de taille supérieure ou égale à 2 car il est équivalent de définir à $x$ et $y$ fixés une unique variable de score pour le groupe $g$ plutot que $\# g$ variables de score individuelles pour chaque passager $p$ de $g$.

Ainsi, préférentiellement, on définira :
<br><br> $$\forall (x,y) \in \mathcal{S}, \forall g \in \mathcal{G}_{2}^{+} : s_{xyg} = c_{côté} ( \delta_{x,y+1,g} + \delta_{x,y-1,g}) 
+ c_{devant} (\delta_{x+1,y,g} + \delta_{x-1,y,g})$$

#### 3.4.2. Sens de parcours de l'avion

De plus, on peut encore diviser par deux le nombre de relations entre les $s_{xyg}$ et les $\delta_{xyg}$ en comptant les <i>liaisons</i> entre passagers plutôt qu'en étudiant l'intégralité du voisinage de ces derniers. En effet, avec la formulation actuelle, nous regardons pour chaque siège si les sièges qui sont devant, derrière, à gauche et à droite de lui sont occupés par des membres du même groupe. Or, remarquons qu'il est strictement équivalent de dire "<i>A est à gauche de B <strong>ET</strong> B est à droite de A</i>" (ce que nous faisons concrètement ici) et de dire "<i>A et B sont à coté l'un de l'autre</i>".

Dès lors, au lieu de compter deux fois chaque liaison, on peut se contenter, pour chaque siège, de regarder les siège à sa droite et derrière lui. On parcourt donc les voisins dans le sens de parcours de l'avion au lieu de revenir en arrière. Ainsi, de manière totalement équivalente à la précédente, on obtient la définition finale des $s_{xyg}$ :
<br><br> $$\forall (x,y) \in \mathcal{S}, \forall g \in \mathcal{G}_{2}^{+} : s_{xyg} = c_{côté} \delta_{x,y+1,g}
+ c_{derrière} \delta_{x+1,y,g}$$

### 3.4. Utilisation des variables $s_{xyp}$

Les variables $s_{xyp}$ sont donc très utiles pour calculer le bonheur des passagers en fonction de leur voisinnage. En revanche, comme nous l'avons expliqué précédemment, si aucun passager n'est assis sur le siège $(x,y)$, la notion de voisinnage perd de son sens. Dès lors, il faut prendre en compte un nouveau type de variables de décision que nous allons appeler les variables $\mu_{xyg}=\mbox{min} \; (M\delta_{xyg},s_{xyg})$, où $M$ est un majorant de l'ensemble des $s_{xyg}$ ($M > 1.5$).

C'est donc sur ces variables $\mu_{xyg}$ que va porter notre fonction objectif. En effet, elles mesurent bien le bonheur des passagers en fonction des passagers assis autour d'eux (ce que l'on souhaite maximiser) en tenant compte de la présence effective ou non des passagers sur es sièges considérés.

### 3.5. Fonction objectif

Ainsi, notre fonction objectif s'écrit : 
$$\boxed {\mbox{Maximize} \;\;\; \sum_{(x,y) \in \mathcal{S}} \sum_{g \in \mathcal{G}_{2}^{+}} \mu_{xyp}}$$

### 3.6. Ecriture complète du problème

$$\mbox{Maximize} \;\;\; \sum_{(x,y) \in \mathcal{S}} \sum_{g \in \mathcal{G}_{2}^{+}} \mu_{xyp}$$
<br><br>
$$\mbox{s.t.} \;\;\; \left\{
    \begin{array}{ll}
        \sum_{p \in \mathcal{P}} \delta_{xyp} \leq 1 & \forall (x,y) \in \mathcal{S} \\
        \sum_{(x,y) \in \mathcal{S}} \delta_{xyp} \leq 1 & \forall p \in \mathcal{P} \\
        \sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}} \delta_{xyp} = \# \mathcal{P} \\
        13 \sum_{p \in \mathcal{P}} m_p \leq \sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}} m_p x \delta_{xyp} \\
        \sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}} m_p x \delta_{xyp} \leq 17 \sum_{p \in \mathcal{P}} m_p \\
        3 \sum_{p \in \mathcal{P}} m_p \leq \sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}} m_p y \delta_{xyp} \\
        \sum_{p \in \mathcal{P}} \sum_{(x,y) \in \mathcal{S}} m_p y \delta_{xyp} \leq 5 \sum_{p \in \mathcal{P}} m_p \\
        \sum_{(x,y) \in \mathcal{S}} x \delta_{xyp} \leq 15 & \forall p \in \mathcal{P}_{90}^{\tau} \\
        \sum_{(x,y) \in \mathcal{S}} x \delta_{xyp} \leq 20 & \forall p \in \mathcal{P}_{120}^{\tau}  \\
        \delta_{xyg} = \sum_{p \in g} \delta_{xyp} & \forall (x,y) \in \mathcal{S}, \forall g \in \mathcal{G}_{2}^{+} \\
        s_{xyg} = c_{côté}^{y} \delta_{x,y+1,g} + c_{derrière}^{x,\# g} \delta_{x+1,y,g} & \forall (x,y) \in \mathcal{S}, \forall g \in \mathcal{G}_{2}^{+} \\
        \mu_{xyg}=\mbox{min} \; (M\delta_{xyg},s_{xyg}) & \forall (x,y) \in \mathcal{S}, \forall g \in \mathcal{G}_{2}^{+}, M > 1.5 \\
        \delta_{xyp} \in \lbrace 0, 1 \rbrace & \forall (x,y) \in \mathcal{S}, \forall p \in \mathcal{P} \\
    \end{array}
\right.$$

### 4.2. Ensemble des passagers

Afin de simplifier l'utilisation des différentes données relatives aux passagers (âge, sexe, poids, classe, temps de correspondance, ...), nous avons fait le choix d'implémenter une classe objet Python pour les modéliser informatiquement. Cela permettra un appel à ces données plus aisé et commode. La classe ```Passager``` est définie ci-dessous.

````python
numPassager = 0
numFemme = 0
numHomme = 0
numEnfant = 0
numWHCR = 0
````

````python
class Passager:
    
    """
    Une classe représentant un passager et toutes les données qui lui sont associées. 
    """
    
    def __init__(self, idx, id_passager, categorie, classe, transit_time=0):
        
        """
        Constructeur pour la classe Passager.
        
        Args:
            idx (int): Indice du groupe dans le fichier Excel
            categorie (Categories): Catégorie du passager (Femme, Homme, Enfant, Fauteil Roulant)
            classe (Classes): Lettre donnant la classe du passager
            transit_time (string, optional): String au format %H:%M:%S donnat le
                temps de transit pour le passager. Defaults to 0.
        """
        
        self.idx = idx
        self.id_passager = id_passager
        self.categorie = categorie
        self.classe = classe
        self.transit_time = transit_time
        self.poids()
    
    def __str__(self):
        return f'Passager # {self.id_passager}, catégorie {self.categorie} du groupe #{self.idx}, classe {self.classe} et temps de correspondance de {self.transit_time}'
    
    def __repr__(self):
        return f'passager # {self.id_passager} du groupe #{self.idx}'          
    
    def poids(self):
        """
        Définit le poids du passager selon sa catégorie.
        """
        
        global numFemme, numHomme, numWHCR, numEnfant
        
        if self.categorie == "femmes":
            self.poids = 65 + numFemme / 100
            numFemme += 1
        elif self.categorie == "hommes":
            self.poids = 80 + numHomme / 100
            numHomme += 1
        elif self.categorie == "WHCR":
            self.poids = 70 + numWHCR / 100
            numWHCR += 1
        else:
            self.poids = 35 + numEnfant / 100
            numEnfant += 1
````

Dans cette même logique, on définit une classe ```Groupe``` pour décrire les groupes de passagers voyageant ensemble.

````python
class Groupe:
    
    """
    Une classe représentant un groupe de passagers ayant réservé leur places d'avion ensemble. 
    Un groupe correspond donc à une ligne dans le fichier Excel de départ.
    """

    def __init__(self, idx, nb_femmes, nb_hommes, nb_enfants, nb_WCHR, classe, transit_time=0):
        
        """
        Constructeur pour la classe Groupe.
        
        Args:
            idx (int): Indice du groupe dans le fichier Excel
            nb_femmes (int): Nombre de femmes dans le groupe
            nb_hommes (int): Nombre d'hommes dans le groupe
            nb_enfants (int): Nombre d'enfants dans le groupe
            nb_WCHR (int): Nombre de personnes handicapés dans le groupe
            classe (Classes): Lettre donnant la classe du groupe
            transit_time (string, optional): String au format %H:%M:%S donnat le
                temps de transit pour le groupe. Defaults to 0.
        """
        
        self.idx = idx
        self.composition = {
            'femmes': nb_femmes,
            'hommes': nb_hommes,
            'enfants': nb_enfants,
            'WHCR': nb_WCHR
        }
        self.classe = classe
        self.transit_time = transit_time

        self.list_passagers = []
        global numPassager

        for categorie in self.composition.keys():
            for nbParCategorie in range(self.composition[categorie]):
                self.list_passagers.append(Passager(
                    self.idx,
                    numPassager,
                    categorie,
                    self.classe,
                    self.transit_time))
                numPassager += 1   

        return

    def __str__(self):
        return f'Groupe #{self.idx} avec {self.get_nombre_passagers()} passager(s), classe {self.classe} et temps de correspondance de {self.transit_time}'

    def __repr__(self):
        return f'groupe #{self.idx}'
    
    def iter_passagers(self):
        """Générateur pour la liste des passagers dans le groupe.
        Yields:
            Passager: instance de la classe Passager
        """
        for passager in self.list_passagers:
            yield passager

    def est_seul(self):
        """Renvoie True si le groupe contient un seul
        passager et False sinon.
        """
        return len(self.list_passagers) == 1
    
    def get_nombre_passagers(self):
        """Renvoie le nombre de passagers dans le groupe considéré.
        """
        return len(self.list_passagers)

    def comprend_enfants(self):
        """Renvoie True si le groupe contient au moins un enfant
        en son sein et False sinon.
        """
        return any([passager.categorie == 'enfants' for passager in self.list_passagers])      
````

### Variables de décision

#### Variables binaires $\pi_{xyp}$

<div style="text-align: justify">
Le problème du placement des passagers est un problème typique d'affectation. Cette catégorie de problèmes d'optimisation a communément recours à des variables binaires commes <i>briques de base</i> afin de représenter numériquement lesdites affectations.

### Exactement un siège par passager

<div style="text-align: justify">
Un autre prérequis est que tous les passagers ayant acheté un billet puissent avoir un siège. Mathématiquement, cette contrainte s'écrit :
<br><br>
$$\forall p \in \mathcal{P}, \; \sum_{(x,y) \in \mathcal{S}_{\emptyset}} \pi_{xyp} = 1$$
<br><br>
Toutefois, après différents tests réalisés avec Gurobi, il s'est avéré qu'implémenter directement l'égalité n'était pas bénéfique pour le modèle car une égalité est une doubles contrainte (un $\leq$ et un $\geq$) ce qui conduit donc à un total de $2\# \mathcal{P}$ contraintes pour exprimer le fait que chaque passager ait un et un seul siège.
<br><br>
Pour outrepasser cet écueil, nous avons fait le choix de casser cette contrainte en deux sous-contraintes mais conduisant au même résultat en moins de contrainte et donc en moins de temps de calcul :
<ul>

    <li> Une seconde contrainte stipulant que le nombre de passagers assis dans l'avion est exactement égal à $\# \mathcal{P}$ (2 contraintes).
</ul>
L'implémentation de ces deux contraintes est détaillée subséquemment.
</div>

### Pas d'enfant isolé

<div style="text-align: justify">
De plus, les enfants voyageant au sein d'un groupe de voyageurs ne doivent pas se retrouver isolés de leur groupe, c'est-à-dire qu'ils doivent nécessairement être assis à côté d'au moins un membre de leur groupe. Afin de formaliser mathématiquement cette contrainte considérons un passager enfant $e \in  \mathcal{P}_{\mathbf{E}}$ tel qu'$e$ fasse partie d'un groupe que l'on note $g_e \in \mathcal{G}_{2}^{+}$. Dès lors, pour tout siège $(x,y) \in \mathcal{S}$ il y a deux cas à considérer :
<ul>
    <li> Soit l'enfant n'est pas assis sur le siège, donc $\pi_{xye} = 0$, et dans ce cas il n'y a rien de particulier à dire;
    <li> Soit l'enfant est effectivement assis sur le siège, donc $\pi_{xye} = 1$, et dans ce cas, la traduction mathématique de la contrainte s'écrit $\gamma_{x,y-1,g_e}+\gamma_{x,y+1,g_e} \geq 1$ (avec des mots : <i>il y a au moins un membre du groupe de $e$ à gauche et/ou à droite de ce dernier</i>).
</ul>
On constate donc que ces deux cas peuvent fusionner en une seule formule qui est la suivante :
<br><br>
$$\gamma_{x,y-1,g_e}+\gamma_{x,y+1,g_e} \geq \pi_{xye}$$
<br>
En effet, soi $\pi_{xye} = 1$ auquel cas on retrouve bien le second cas, soi $\pi_{xye} = 0 \Rightarrow \gamma_{x,y-1,g_e}+\gamma_{x,y+1,g_e} \geq 0$ ce qui n'apporte aucune restriction supplémentaire.
<br><br>
Ainsi, dans le cas général, la contrainte s'écrit :
<br><br>
$$\forall (x,y) \in \mathcal{S}, \; \forall g \in \mathcal{G}_{2}^{+}, \; \forall e \in g \cap \mathcal{P}_{\mathbf{E}}, \; \gamma_{x,y-1,g}+\gamma_{x,y+1,g} \geq \pi_{xye}$$
<br><br>
Un enfant pourra toujours être à côté un adulte tant que $\#E \leq 2\#A$
</div>

````python
placement = {}

for passager in listePassagers:
    for x,y in avion['seats']['real']:
        if PI[x,y,passager.id_passager].x == 1:
            placement[passager.id_passager] = (x,y)  
            
for groupe in listeGroupes.keys():
    print("Groupe #{} (temps de correspondance : {} min):".format(groupe,listeGroupes[groupe].transit_time))
    for passager in listeGroupes[groupe].list_passagers:
        if passager.id_passager in placement.keys():
            coord = placement[passager.id_passager]
            x,y = coord
            print("{} assis sur le siège {} = {}".format(passager.id_passager,coord,coordToSiege(x,y)))
    print("\n")    
````

### Permutation des groupes

````python
for groupe in listeGroupes:
    if listeGroupes[groupe].get_nombre_passagers() >= 4:
        x_min = min([placement[passager.id_passager][0] for passager in listeGroupes[groupe].list_passagers])
        x_max = max([placement[passager.id_passager][0] for passager in listeGroupes[groupe].list_passagers])
        y_min = min([placement[passager.id_passager][1] for passager in listeGroupes[groupe].list_passagers])
        y_max = max([placement[passager.id_passager][1] for passager in listeGroupes[groupe].list_passagers])
        print(x_min,x_max,y_min,y_max)
        cluster = [(i,j) for i in range(x_min,x_max+1) for j in range(y_min,y_max+1)]
        corners = [(x_min,y_min),(x_min,y_max),(x_max,y_min),(x_max,y_max)]
        
        if listeGroupes[groupe].get_nombre_passagers() != 13: # On gère à part le cas pathologique
            positions = {}
            void_coord = (0,0)
            for i,j in cluster:
                if (i,j) in avion['seats']['fictive'] or sum((passager.idx+1)*PI[i,j,passager.id_passager].x for passager in listePassagers) == groupe + 1:
                    positions[i,j] = 1
                elif sum((passager.idx+1)*PI[i,j,passager.id_passager].x for passager in listePassagers) == 0:
                    positions[i,j] = 0
                    void_coord = (i,j)
                else:
                    positions[i,j] = -1
                    void_coord = (i,j)
            print(positions)                        
            if void_coord not in corners and void_coord != (0,0):
                i,j = void_coord
                print("hello")
                if positions[i,j] == 0: # S'il y a un vide entre les passagers du groupe
                    if (i,j) in avion['seats']['exit']: # Si le trou est au niveau d'une issue de secours
                        if j <= 4:
                            # On échange avec le passager tout à gauche de la rangée
                            swap_passager = sum((passager.id_passager)*PI[i,y_min,passager.id_passager].x for passager in listePassagers)
                            placement[int(swap_passager)] = (i,j)

                        else:
                            # On échange avec le passager tout à droite de la rangée
                            swap_passager = sum((passager.id_passager)*PI[i,y_max,passager.id_passager].x for passager in listePassagers)
                            placement[int(swap_passager)] = (i,j)
                            
                    else: # Sinon dans le cas général
                        # On échange avec le passager tout en bas à gauche
                        swap_passager = sum((passager.id_passager)*PI[x_max,y_min,passager.id_passager].x for passager in listePassagers)
                        print(swap_passager)
                        placement[int(swap_passager)] = (i,j)
                        
                elif positions[i,j] == 1: # S'il y a un passager qui n'est pas membre du groupe dans le cluster  
                    if (i,j) in avion['seats']['exit'] or i == x_min: # Si le trou est au niveau d'une issue de secours ou que le passager a un temps de correspondance inférieur à celui du groupe
                        stranger = sum((passager.id_passager)*PI[i,j,passager.id_passager].x for passager in listePassagers)
                        swap_passager = sum((passager.id_passager)*PI[i,y_min,passager.id_passager].x for passager in listePassagers)
                        placement[int(swap_passager)] = (i,j)
                        placement[int(stranger)] = (i,y_min)
                    else: # Sinon dans le cas général
                        stranger = sum((passager.id_passager)*PI[i,j,passager.id_passager].x for passager in listePassagers)
                        swap_passager = sum((passager.id_passager)*PI[x_max,y_min,passager.id_passager].x for passager in listePassagers)
                        placement[int(swap_passager)] = (i,j)
                        placement[int(stranger)] = (x_max,y_min)
                        
                    
                    
            
            
````

### Export du résultat en CSV

````python
def exporter_resultats(listeGroupes, placement):    
    list_id_groupe = []
    list_id_passager = []
    list_categorie = []
    list_classe = []
    list_transit_time = []
    list_poids = []
    list_x = []
    list_y = []
    list_siege = []
    list_classes = []
    
    for groupe in listeGroupes.keys():        
        for passager in listeGroupes[groupe].list_passagers:
            if passager.id_passager in placement.keys():
                x, y = placement[passager.id_passager]
                siege = coordToSiege(x, y)
                
                list_id_groupe.append(groupe)
                list_id_passager.append(passager.id_passager)
                list_categorie.append(passager.categorie)
                list_classe.append(passager.classe)
                list_transit_time.append(passager.transit_time)
                list_poids.append(passager.poids)
                list_x.append(x)
                list_y.append(y)
                list_siege.append(siege)
                list_classes.append(get_classe(x, y))
    
    df = pd.DataFrame({
        'ID Groupe': list_id_groupe,
        'ID Passager': list_id_passager,
        'Catégorie': list_categorie,
        'Classe': list_classe,
        'Transit Time': list_transit_time,
        'Poids': list_poids,
        'x': list_x,
        'y': list_y,
        'Siège': list_siege,
        'Classe': list_classes
    })
    
    return df
````
